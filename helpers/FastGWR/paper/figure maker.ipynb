```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
gwr = np.array([18.0,31.07,57.89,107.72,210.04,673.0])

mgwr = np.array([888.39,1165.61,1735.79, 2762.86, 4825.57, 22451.34])

mgwr_2 = np.array([163.48,317.79,621.77,1243.5, 2429.0, 12047])

mgwr_3 = mgwr - mgwr_2

nproc = [80,40,20,10,5,1]
```

```python
gwr = gwr[-1]/gwr
mgwr = mgwr[-1]/mgwr
mgwr_2 = mgwr_2[-1]/mgwr_2
mgwr_3 = mgwr_3[-1]/mgwr_3

flg = plt.figure(figsize=(6, 6))

plt.plot(nproc,gwr,marker="o",label="GWR (total)",markersize=5)
plt.plot(nproc,mgwr,marker="o",label="MGWR (total)",markersize=5)
plt.plot(nproc,mgwr_2,marker="o",label="MGWR (estimation only)",markersize=5)
#plt.plot(nproc,mgwr_3,marker="o",label="MGWR (estimation only)",markersize=5)
plt.ylim(-3,83)
plt.ylabel('Speed-up',fontsize=13)
plt.xlabel('Number of Processors',fontsize=13)
plt.legend(loc='upper left',fontsize=13)
#plt.title('(b)',pad=-390)

plt.tight_layout()
plt.savefig('scalability.png', format='png', dpi=900)
```
